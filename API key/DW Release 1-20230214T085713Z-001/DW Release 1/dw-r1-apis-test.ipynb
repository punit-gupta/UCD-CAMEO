```python
import requests
import json
```

## Search API
Search for datasets that meet specified criteria.

```python
url = "http://10.11.1.6:8000/cameo/search"
params = {
    "footprint": "3.46191 6.16042, -4.59641 37.9012, -6.697357 54.65898353711527",
    "platformname":"Sentinel-2",
    "fromDate":"2022-01-01",
    "username": "user1"
}
response = requests.get(url=url, json=params)

if response.status_code ==200:
    print(json.dumps(response.json(), indent=2))
else:
    print(response.status_code)
```

## Metadata retrieval API

```python
url = "http://10.11.1.6:8000/cameo/metadata/1dfcbac9-c227-4b9b-afa7-6b6f9e68206a"
params = {
    "username": "user1"
}
response = requests.get(url=url, json=params)

if response.status_code ==200:
    print(json.dumps(response.json(), indent=2))
else:
    print(response.status_code)
```

## Data retrieval API
Retrieve the name of the file that contains the specified dataset.

```python
url = "http://10.11.1.6:8000/cameo/data/1dfcbac9-c227-4b9b-afa7-6b6f9e68206a"
params = {
    "username": "user1"
}
response = requests.get(url=url, json=params)

if response.status_code ==200:
    print(json.dumps(response.json(), indent=2))
else:
    print(response.status_code)
```
