```python
S_sentinel_bands = glob("IMG_DATA/*B?*.jp2")
S_sentinel_bands.sort()
print(S_sentinel_bands)
l = []
for i in S_sentinel_bands:
    with rio.open(i, 'r') as f:
        l.append(f.read(1, window = Window(0, 0, 1800, 1800)))
        
print(l)
```
